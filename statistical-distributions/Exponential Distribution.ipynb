```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

<i>The Exponential distribution</i> is a continuous exponential distribution which infers <b>the probability of the waiting time between events</b> in a Poisson distribution process.<br>

The Exponential distribution is a part of the large class of probability distibutions which also includes Binomial, Gamma and Poisson distributions.

### Poisson Distribution Process


The Poisson process is used to describe a counting process where the events happen at random but at a given rate.<br>

The Poisson Process is defined by <b>the Poisson Distribution Formula (PMF)</b>:<br>
##### $P(k\space events\space in\space interval) = e^{-\lambda}\frac{\lambda^k}{k!}$
- $\lambda$ is an average number of events per interval
- $e$ is Euler's number
- $k$ is a random value parameter which may take integer values (0, 1, 2...)
    - $k$ is a number of events occurring within a certain time period




## PDF - Probability Density Function
> PDF is used to describe the continuous probability distributions

#### Formula

#### $f(x,\lambda)=\lambda e^{-\lambda\tau}$
- $\lambda$ is an average number of events per interval
- $e$ is Euler's number
- $\tau$ is a time interval


```python
x = np.arange(0, 10, .1)

plt.title('PDF - Probability Density Function')
plt.plot(
    x,
    stats.expon.pdf(x=x), 
    label='scale: 1'
)
plt.plot(
    x,
    stats.expon.pdf(x=x, scale=2), 
    label='scale: 2'
)
plt.xlabel('x')
plt.ylabel('P(x)')
plt.legend()

plt.show();
```


## RVS - Random variates


```python
plt.hist(
    stats.expon.rvs(size=1000, random_state=12), 
    density=True,
    bins='auto'
)
plt.xlabel('x')
plt.ylabel('P(x)')

plt.show();
```


## PPF - Percent Point Function
- Inverse of $cdf$
- Returns a data point from a declared percentile

```python
x = np.arange(0, 5, .01)
quantile_list = [.1, .5, .9, .99]

plt.title('Exponential Distribution PDF and look for quantiles data points')
plt.plot(
    x,
    stats.expon.pdf(x=x),
    color='darkorange',
    linewidth=3,
    alpha=.6
)
# Quantiles
for quantile in quantile_list:
    data_point = stats.expon.ppf(q=quantile)
    plt.vlines(
        x=data_point, 
        ymin=0,
        ymax=1,
        color='cadetblue', 
        alpha=.8, 
        linestyle=(0, (5, 10)),
        # label=f'quantile: {quantile}'
    )
    plt.annotate(f'x: {round(data_point, 2)}', xy=(data_point + .03, 0.95), color='darkslategrey', alpha=.9, fontsize=8)
    plt.annotate(f'q: {quantile}', xy=(data_point + .03, 0.92), color='darkslategrey', alpha=.9, fontsize=8)

plt.tick_params(labelsize=6)
plt.xlabel('x')
plt.ylabel('P(x)')

plt.show();
```
