```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# set mean and standard difference for average people's height
mu = 172
sd = 9
```

### RVS - Random variates

```python
# generate samples from our distribution
x = stats.norm.rvs(loc=mu, scale=sd, size=1000)
```

```python
plt.hist(
    x,
    bins='auto',
    histtype='stepfilled'
)
plt.vlines(mu, 0, 120, linestyle="--", color='red', label='mu setted')
plt.title("Average People's Height Normal Distribution");
```

```python
# Some stats from the distribution
print('Maximum Likelihood Mean: ', round(x.mean(),3))
print('Maximum Likelihood Median: ', round(np.median(x),3))
print('Maximum Likelihood Variance: ', round(x.var(),3))
print('Maximum likelihood Std: ', round(x.std(),3))
print('Unbiased Variance: ', round(x.var(ddof=1),3))
print('Unbiased Std: ', round(x.std(ddof=1),3))
```


### PDF - Probability Density Function

#### Formula

#### $f(x) = \frac{1}{\sigma\sqrt{2\pi}}exp(-\frac{{(x-u)}^2}{2{\sigma}^2})$
> [How the formula works](https://borisagain.substack.com/p/understanding-the-normal-distribution)

- The default distribution is in standardized form, where formula is:
$f(x) = \frac{\exp({-x^2}/2)}{\sqrt{2\pi}}$

- To change the distribution by shift or scale parameters an equivalent `norm.pdf(y) / scale` is used, where `y = (x - loc) / scale`

```python
x = np.linspace(stats.norm.ppf(0.01), stats.norm.ppf(0.99), 1000)
print(x.min(), x.max(), x.mean())
```

```python
plt.plot(x, stats.norm.pdf(x),
         'r-', lw=5, alpha=0.6, label='norm pdf')
plt.title('Probability Density Function');
```

### PPF - Percent Point Function
- Inverse of $cdf$
- Returns a data point from a declared percentile

```python
print('What is the height of people from 10th percentile? ', 
      round(stats.norm.ppf(0.1, loc=mu, scale=sd),1))
print('What is the height of people from 50th percentile (median)? ', 
      round(stats.norm.ppf(0.5, loc=mu, scale=sd),1))
print('What is the height of people from 90th percentile? ', 
      round(stats.norm.ppf(0.9, loc=mu, scale=sd),1))
print('What is the height of people from 95th percentile? ', 
      round(stats.norm.ppf(0.95, loc=mu, scale=sd),1))
print('What is the height of people from 99th percentile? ', 
      round(stats.norm.ppf(0.99, loc=mu, scale=sd),1))
```

### CDF - Cumulative Distribution Function
- Inverse of $ppf$
- Returns a percentile from a declared data point

```python
print('What is the percentile for people with 160cm height? ', 
      round(stats.norm.cdf(160, loc=mu, scale=sd),2))
print('What is the percentile for people with 172cm height? ', 
      round(stats.norm.cdf(172, loc=mu, scale=sd),2))
print('What is the percentile for people with 187cm height? ', 
      round(stats.norm.cdf(187, loc=mu, scale=sd),2))
print('What is the percentile for people with 193cm height? ', 
      round(stats.norm.cdf(193, loc=mu, scale=sd),2))
```

### SF - Survival Function
- Can be defined as 1 - $cdf$

```python
print('If you are 183cm height, what is the possibility that someone is taller then you? ', 
      round(stats.norm.sf(183, loc=mu, scale=sd), 4))
print('If you are 183cm height, what is the possibility that someone is taller then you? ', 
      round(1 - stats.norm.cdf(183, loc=mu, scale=sd), 4))
```
